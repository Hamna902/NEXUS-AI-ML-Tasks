# Task 6: Forecasting Future Sales (Time Series Analysis)

### Step1. 
**Data Preparation:**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
import numpy as np
from datetime import timedelta

# Step 1: Data Preparation
# Load the dataset (assuming it's in the current directory or provide the full path)
df = pd.read_csv('train.csv')

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')

# Aggregate sales by date (sum sales for each day)
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()
daily_sales.set_index('Order Date', inplace=True)

# Resample to daily frequency (fill missing dates with 0 if no sales)
daily_sales = daily_sales.asfreq('D', fill_value=0)

# Check for stationarity using Augmented Dickey-Fuller test
def check_stationarity(timeseries):
    result = adfuller(timeseries)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    if result[1] > 0.05:
        print("Series is not stationary. Applying differencing...")
        return timeseries.diff().dropna()
    else:
        print("Series is stationary.")
        return timeseries

# Make the series stationary if needed
stationary_sales = check_stationarity(daily_sales['Sales'])

# For SARIMA, we might need seasonal differencing as well. 
# Assuming a seasonal period of 7 days (weekly seasonality for sales data)
# If not stationary after first differencing, apply seasonal differencing
if adfuller(stationary_sales)[1] > 0.05:
    stationary_sales = stationary_sales.diff(7).dropna()  # Seasonal differencing
```

### Step 2. 
**Model Training**


```python
# Step 2: Model Training
# Split data into train and test (e.g., last 30 days as test set)
train = daily_sales[:-30]
test = daily_sales[-30:]

# Train SARIMA model
# Parameters: (p,d,q) for ARIMA, (P,D,Q,s) for seasonal. Here, using example values.
# In practice, use auto_arima from pmdarima or grid search to find best params.
# For simplicity, assuming (1,1,1) x (1,1,1,7) - adjust based on your data/ACF/PACF plots.
model = SARIMAX(train['Sales'], order=(1,1,1), seasonal_order=(1,1,1,7))
model_fit = model.fit(disp=False)
```

### Step 3. 
**Forecasting**

```python
# Step 3: Forecasting
# Predict on test set
pred = model_fit.get_prediction(start=test.index[0], end=test.index[-1])
pred_mean = pred.predicted_mean
pred_conf = pred.conf_int()

# Forecast future 30 days
future_dates = [daily_sales.index[-1] + timedelta(days=i) for i in range(1, 31)]
future_forecast = model_fit.get_forecast(steps=30)
future_mean = future_forecast.predicted_mean
future_conf = future_forecast.conf_int()

# Create DataFrames for forecasts
pred_df = pd.DataFrame({'Predicted Sales': pred_mean}, index=test.index)
future_df = pd.DataFrame({'Forecasted Sales': future_mean}, index=future_dates)
```

### Step 4. 
**Visualization**

```python
# Step 4: Visualization
plt.figure(figsize=(14, 7))
plt.plot(daily_sales['Sales'], label='Historical Sales')
plt.plot(pred_df['Predicted Sales'], label='Test Set Predictions')
plt.plot(future_df['Forecasted Sales'], label='Future Forecast')
plt.fill_between(future_conf.index, future_conf.iloc[:, 0], future_conf.iloc[:, 1], color='k', alpha=0.1)
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Print the future forecast
print("Future 30-Day Sales Forecast:")
print(future_df)
```
